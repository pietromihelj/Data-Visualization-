```python
import pandas as pd
df = pd.read_csv("assignment_3.csv")  
df.head()
```

```python
usage = df.iloc[2:,162:198]
freaquency = df.iloc[2:,199:235]
likelyness = df.iloc[2:, 236:272]
```

```python
df_final = pd.concat([usage, freaquency, likelyness], axis=1)
df_final.head()
```

```python
df_final.drop(df.index[3], inplace=True)
df_final.head()
```

```python
df_final.columns = df_final.iloc[0,:]
df_final = df_final[1:].reset_index(drop=True)
df_final.head()
```

```python
usage = df_final.iloc[:,:36]
freaquency = df_final.iloc[:,36:72]
likelyness = df_final.iloc[:,72:]
```

```python
import numpy as np
from collections import defaultdict
```

```python
use = defaultdict()
for colonna in usage.columns:
    use[colonna] = usage[colonna].value_counts()
print(use)
```

```python
freq = defaultdict()
for colonna in freaquency.columns:
    freq[colonna] = freaquency[colonna].value_counts()
print(freq)
```

```python
like = defaultdict()
for colonna in likelyness.columns:
    like[colonna] = likelyness[colonna].value_counts()
print(like)
```

```python
del use['Other (please specify)'] 
del freq['[Other]'] 
del like['[Other]'] 
```

```python
import matplotlib.pyplot as plt

categories = list(use.keys())
use_values = []
for name in categories:
    use_values.append(getattr(use[name],name))

sindex = np.argsort(use_values)
use_values = np.sort(use_values)
categories = [categories[i] for i in sindex]
mean = np.mean(use_values)

fig = plt.figure(figsize=(10,10))
plt.barh(categories[:23], use_values[:23], color = 'red')
plt.barh(categories[23:], use_values[23:], color = 'blue')
plt.axvline(x=mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')  # Linea della media
plt.title('Usage')
plt.xlabel('Tools')
plt.ylabel('Numbers of people using it')
plt.show()
```

```python

attr = ['Rarely', 'Sometimes', 'Often']
categorie = freq.keys()
freq_abs_values = []
for name in categorie:
    freq_abs_values.append(max(attr, key=lambda a: getattr(freq[name], a)))

mapping = {'Rarely': 20, 'Sometimes': 100, 'Often': 400}
freq_abs_values = [mapping[val] if val in mapping else val for val in freq_abs_values]
freq_abs_values = [freq_abs_values[i] for i in sindex]

like['Svelte']['Not at all'] = 0
like['Physical materials (other than pen and paper)']['Not at all'] = 0
attr = ['Very much', 'Somewhat', 'Not at all']
categorie = like.keys()
like_abs_values = []
for name in categorie:
    like_abs_values.append(max(attr, key=lambda a: getattr(like[name], a)))
like_abs_values = [like_abs_values[i] for i in sindex]

colors = []
for a in like_abs_values:
    if a == 'Not at all':
        colors.append('#93003a')
    if a == 'Somewhat':
        colors.append('#7c7c7c')
    else:
        colors.append('#003a93')

print(len(freq_abs_values))
print(len(use_values))
print(len(like_abs_values))
print(len(colors))
```

```python
c = sum(use_values[:22])
use_values = use_values[23:]
use_values = np.insert(use_values,0,c)
print(use_values)
from collections import Counter
temp = freq_abs_values[:22]
counter = Counter(temp)
M, count = counter.most_common(1)[0]
freq_abs_values = freq_abs_values[23:]
freq_abs_values.insert(0,M)
print(freq_abs_values)
from collections import Counter
temp = like_abs_values[:22]
counter = Counter(temp)
G, count = counter.most_common(1)[0]
like_abs_values = like_abs_values[23:]
like_abs_values.insert(0,G)
categories = list(categories)
categories = categories[23:]
categories.insert(0,'Oters')
print(len(categories))
print(len(use_values))
print(len(like_abs_values))
print(len(colors))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.lines import Line2D
fig, ax = plt.subplots(1, 1, figsize=(15, 14))
categories = list(categories)
# Crea il grafico a dispersione
scatter = ax.scatter(use_values[1:22], categories[1:22], s=freq_abs_values[1:22], alpha=0.5, c=colors[1:22], edgecolors="gray")
scatter = ax.scatter(use_values[23:], categories[23:], s=freq_abs_values[23:], alpha=1, c=colors[23:], edgecolors="black")
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
ax.set_xlabel("Number of people using it")
ax.set_ylabel("Tools")
ax.set_title("Popularity Of Visualization Tools")
ax.grid(True, which='both', axis='y', linestyle='-', linewidth=0.7, alpha=0.5)
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#93003a', markersize=10, label='Not liked'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#7c7c7c', markersize=10, label='Indifferent'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#003a93', markersize=10, label='Liked')
]
legend_elements_sizes = [
    Line2D([0], [0], marker='o', color='w', markersize=10, label='Rarely used', markerfacecolor='#7c7c7c'),
    Line2D([0], [0], marker='o', color='w', markersize=15, label='Sometimes used', markerfacecolor='#7c7c7c'),
    Line2D([0], [0], marker='o', color='w', markersize=20, label='Often used', markerfacecolor='#7c7c7c')
]
ax.legend(handles=legend_elements+legend_elements_sizes, loc='upper left', title='Likeliness', fontsize=12, bbox_to_anchor=(1.05, 1), borderpad=1)
plt.tight_layout()
plt.show()
```
